# API Working With Requests package

```python
import requests
```

```python
def fetch_content_api():
  url = "https://api.freeapi.app/api/v1/public/randomusers/user/random"#add url of API
  response = requests.get(url)#if GOOD Response then 200 and 'GET' data
  data = response.json() # Save the response as JSON
  if data ["success"] and "data" in data: #Check if data is available
    user_data = data["data"] #data is another section inside the JSON
    user_username = user_data["login"]["username"] #DATA ----> LOGIN ------> USERNAME
    user_country = user_data["location"]["country"] #DATA ----> LOCAtion ------> COUNTRY
    return user_username, user_country
  else:
    raise Exception("Failed to fetch User data") #in case of error
```

```python
def main():
  try:
    fetch_content_api()
    print("Content fetched successfully")

  except Exception as e:
    print(f"An error occurred: {e}")
```

```python
if __name__ == "__main__":
  main()
```

# API Working With Requests Try and Catch Block

```python
import requests
```

```python
def fetch_content_using_TC():
  try:
    url = "https://api.freeapi.app/api/v1/public/randomusers/user/random" #add url of API
    response = requests.get(url) #if GOOD Response then 200 and 'GET' data
    data = response.json() # Save the response as JSON
    user_data = data["data"] #data is another section inside the JSON
    user_name= user_data["login"]["username"] #DATA ----> LOGIN ------> USERNAME
    user_country = user_data["location"]["country"] #DATA ----> LOCAtion ------> COUNTRY
    return user_name, user_country
  except Exception as e: #in case of error
    print(f"An error occurred: {e}")
```

```python
fetch_content_using_TC()
```

# API Working With Asyncio
`async` functions are designed to `await` tasks that might take time to complete, without blocking the rest of your program. Other parts of the program can continue running while these tasks complete.

`asyncio` is the library that supports this concurrency in Python.

```python
import aiohttp
import asyncio
```

```python
url = "https://api.freeapi.app/api/v1/public/randomusers/user/random"
```

```python
async def fetch_data_Asyncio():
    async with aiohttp.ClientSession() as session:  # Work only till session is valid
        try:
            async with session.get(url) as response:  # Get URL
                print(f"Status Code: {response.status}")  # Print the status code
                response.raise_for_status()  # Check if the request was successful

                data = await response.json()  # Parse the JSON data
                print("Raw JSON response:", data)  # Print the entire response for inspection

                user_data = data["data"]  # 'data' is another section inside the JSON

                # Extract login details
                user_name = user_data["login"]["username"]

                # Extract location details
                user_country = user_data["location"]["country"]

                print(f"User: {user_name}, Country: {user_country}")
        except aiohttp.ClientError as err:
            print(f"HTTP error occurred: {err}")  # Error in HTTP
        except Exception as err:
            print(f"An error occurred: {err}")  # Error in fetching
```

```python
await fetch_data_Asyncio()
```

```python
def fetch_products():
  url="https://api.freeapi.app/api/v1/public/randomproducts/product/random/"
  response = requests.get(url)
  data = response.json()
  user_brand = data["data"]["brand"]
  user_images = data["data"]["images"][1]
  user_price = data["data"]["price"]
  return user_brand, user_images , user_price
```
